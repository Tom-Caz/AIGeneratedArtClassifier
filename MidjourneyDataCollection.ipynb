# Midjourney Data Collection

This program downloads the data stored in url format from the Midjourney dataset.

---


```python
import json
import requests
from PIL import Image
```

```python
size = (256,256)
```

```python

with open('Unprocessed Data/Midjourney/general-01_2022_06_20.json', 'r') as myfile:
    data=myfile.read()
obj = json.loads(data)
for i, x in enumerate(obj['messages']):
    
    # If image exists and is upscaled:
    if obj['messages'][i][0]['attachments'] and "Upscaled" in obj['messages'][i][0]['content']:
        image_url = obj['messages'][i][0]['attachments'][0]['url']
        print(image_url)
        
        img_data = requests.get(image_url).content
        file_name = "Unprocessed Data/MidjourneyImages/Midjourney" + obj['messages'][i][0]['id'] + ".jpg"
        with open(file_name, 'wb') as handler:
            handler.write(img_data)
        # print(x)
        im = Image.open(file_name)
        im = im.resize(size)
        im.save(file_name)
        print(i)
        print(im.size)
```
