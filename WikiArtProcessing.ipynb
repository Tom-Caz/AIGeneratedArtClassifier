# WikiArt Data Processing

This program processes the WikiArt figurative painting data

---

```python
from PIL import Image
import os
```

```python
size = (256,256)
```

```python
directory = "Unprocessed Data/WikiArt/images/figurative"
for filename in os.scandir(directory):
    if filename.is_file():
        print(filename.path)
        im = Image.open(filename.path)
        im = im.resize(size)
        im.save("Unprocessed Data/WikiArtFigProcessed/" + filename.name, "JPEG")
        print(im.size)

        
```

```python
directory = "Unprocessed Data/WikiArtFigProcessed"
for i, filename in enumerate(os.scandir(directory)):
    if filename.is_file():
        print(i)
        print(filename.path)
        
```
